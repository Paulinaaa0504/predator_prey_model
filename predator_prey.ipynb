### Wprowadzenie

W matematyce istnieje wiele modeli opisujących zmiany liczebności populacji. W naszych rozważaniach, zamiast skupiać się na pojedynczej jednostce, która żyje i rozmnaża się w pewnym środowisku, będziemy analizować bardziej realistyczny model, w którym nie jest ona sama, lecz egzystuje z innymi gatunkami, które mogą mieć wpływ na tempo w jakim się rozmnaża i wymiera. W rozważanym schemacie będziemy dzielić organizmy na dwie kategorie: drapieżniki i ofiary. Jedna z grup, aby się efektywnie rozmnażać potrzebuje wystarczającej liczby ofiar, która przez to zmniejsza się w czasie. Druga populacja, aby się powiększać potrzebuje, aby liczność drapieżników była mniejsza, żeby te miały większe szanse na przeżycie. Możemy tutaj myśleć o wilkach i królikach w odizolowanym lesie, mszycach i biedronkach lub bakteriach i amebach. Ten model można opisać za pomocą równań różniczkowych. Dzięki niemu możemy przedstawić dynamikę życia organizmów żywych.


Układ takich równań został w 1920 roku przedstawiony przez Alfreda Jamesa Lotka, który stworzył model do opisu oscylacji stężeń substancji w reakacji chemicznej. Niezależnie od tego w 1926 roku włoski matematyk Vito Voltera przedstawił je do opisu zjawiska, które zaszło po I wojnie światowej w Morzu Adriatyckim. Paradoksalnie, mimo trwających działań wojennych, liczba ryb drapieżnych w tym morzu wzrosła. Volterra, korzystając ze swojego modelu, wykazał, że ten wzrost był całkowicie naturalny, ponieważ w czasie wojny połowy ryb zostały wstrzymane, co pozwoliło populacji drapieżników powrócić do swojej naturalnej liczebności. ### Założenia

Przed przedstawieniem modelu Lotki-Voltery musimy zastosować odpowiednie założenia, ponieważ jest on dużym uproszczeniem tego, co się dzieje w rzeczywistości. Zakładamy, że [1]:
* w ekosystemie istnieją jedynie dwa gatunki nazywane drapieżnikami (P) i ofiarami (V), gdzie ofiary stanowią pokarm drapieżników,
* jeśli nie ma drapieżników, to ofiary rozmnażają się zgodnie z modelem Malthusa,
* jeśli nie ma ofiar, drapieżniki umierają z braku pożywienia,
* jeśli występują razem na danym odizolowanym obszarze, część ofiar ginie przy kontakcie z drapieżnikami, dzięki czemu te drugie mogą się rozmnażać,
* gatunki te poruszają się losowo, niezależnie od siebie, dlatego liczba kontaktów (polowań), jednej grupy jest proporcjonalna do liczebności drugiej.


### Model Lotki-Volterry
Model Lotki-Volterry opisywany jest układem równań:
$$
\begin{cases}
    \frac{\mathrm{d}V}{\mathrm{d}t} = rV - aVP, \\[5pt]
    \frac{\mathrm{d}P}{\mathrm{d}t}= -sP + bVP.
\end{cases}
$$

Gdzie $ a, b, r, s > 0 $ (stałe) oraz:
* V - liczba ofiar,
* P - liczba drapieżników,
* r - współczynnik przyrostu ofiar,
* a - skuteczność drapieżników,
* s - współczynnik umierania drapieżników,
* b - współczynnik przyrostu drapieżników.


### Punkty stacjonarne

Aby przeanalizować ten model znajdziemy teraz wartości, dla których liczebność obu populacji nie zmienia się w czasie, czyli punkty stacjonarne naszych równań $\left( V_s, P_s \right)$. Otrzymujemy:
$$
\begin{cases}
\frac{\mathrm{d}V}{\mathrm{d}t} = 0,\\[5pt]
\frac{\mathrm{d}P}{\mathrm{d}t}= 0.
\end{cases}
$$
Rozwiązując te równania otrzymujemy dwa takie punkty:
$$
\left( V_s, P_s \right) = \left( 0, 0 \right) \hspace{0.5cm} \text{albo} \hspace{0.5cm}\left( V_s, P_s \right) = \left( \frac{s}{b}, \frac{r}{a} \right).
$$
Aby przeanalizować stabilność tych punktów stacjonarnych skorzystamy z macierzy Jacobiego:
$$
J\left(V, P\right) = \left[\begin{array}{cc}
\frac{\mathrm{d}\left(rV - aVP\right)}{\mathrm{d}V} & \frac{\mathrm{d}\left(rV - aVP\right)}{\mathrm{d}P}\\[5pt]
\frac{\mathrm{d}\left(-sP + bVP\right)}{\mathrm{d}V} & \frac{\mathrm{d}\left(-sP + bVP\right)}{\mathrm{d}P}
\end{array}\right]
= \left[\begin{array}{cc}
r - aP & -aV\\[5pt]
bP & -s +bV
\end{array}\right]
$$
Dla punktu $\left( V_s, P_s \right) = \left( 0, 0 \right)$ otrzymujemy:
$$
J\left(0, 0\right) =
\left[\begin{array}{cc}
r & 0\\[5pt]
0 & -s
\end{array}\right]
$$
Dostajemy stąd wartości własne $\lambda _1 = r$ i $\lambda _2 = -s$. Widzimy, że $\text{Re}\left(\lambda _1\right) > 0$, więc jest to punkt niestabilny.
Natomiast dla drugiego punktu $\left( V_s, P_s \right) = \left( \frac{s}{b}, \frac{r}{a} \right)$ mamy:
$$
J\left( \frac{s}{b}, \frac{r}{a} \right) =
\left[\begin{array}{cc}
0 & \frac{-as}{b}\\[5pt]
\frac{br}{a} & 0
\end{array}\right]
$$
Otrzymujemy stąd wartości własne $\lambda _{1,2} = \pm i \sqrt{rs}$. Są one czysto urojone, dlatego punkt ten jest neutralnie stabilny, więc układ nie jest hiperboliczny. To oznacza, że punkt $\left( V_s, P_s \right) = \left( \frac{s}{b}, \frac{r}{a} \right)$ jest centrum.


### Portret fazowy
Przeanalizujemy teraz jak liczebność jednej populacji ma wpływ na liczebność drugiej. Z poprzednich rozważań wiemy już, że wartości te będą oscylować wokół punktu $\left( \frac{s}{b}, \frac{r}{a} \right)$.
Znajdziemy teraz izokliny zerowe, czyli krzywe, dla których liczebność poszczególnych populacji nie ulega zmianie:
$$
\frac{\mathrm{d}V}{\mathrm{d}t} = 0  \implies V = 0 \hspace{6pt} \lor \hspace{6pt} V = \frac{s}{b}, \\[5pt]
\frac{\mathrm{d}P}{\mathrm{d}t}= 0 \implies P = 0 \hspace{6pt} \lor \hspace{6pt} P = \frac{r}{a}.
$$
Te izokliny dzielą naszą przestrzeń na cztery grupy, dla których zmienne $V, P$ mają różne kieunki.
Izokliny, punkty stacjonarne i kierunki zmian poszczególnych zmiennych zostały przedstawione na poniższym wykresie [2]:

Powstałe z tego pole wektorowe pokazuje, że rozwiązania tworzą zamknięte orbity, które okrążają punkt stacjonarny $\left( V_s, P_s \right) = \left( \frac{s}{b}, \frac{r}{a} \right)$. Oznacza to, że liczebność ofiar osiąga maksimum tuż przed maksymalną liczebnością drapieżników. \
\
Na wykresie poniżej przedstawione zostało pole wektorowe dla przykładowych parametrów. Długość wektora odpowiada tempu zmiany populacji w danym punkcie, a kierunek wskazuje, w którą stronę zmieniają się liczebności ofiar i drapieżników. Dodatkowo, na wykresie uwzględniono cykle trajektorii dla przykładowych wartości początkowych [2].


```python
from scipy.integrate import odeint

r, a, b, s = 0.1, 0.02, 0.01, 0.1
V_s = s / b
P_s = r / a

def lotka_volterra(X, t):
    V, P = X
    dVdt = r * V - a * V * P
    dPdt = b * V * P - s * P
    return [dVdt, dPdt]


V = np.linspace(0, 25, 20) 
P = np.linspace(0, 12.5, 20)  
V, P = np.meshgrid(V, P)
DV, DP = lotka_volterra([V,P], 0)
M = np.hypot(DV, DP)

initial_conditions = [
    (10, 4), 
    (15, 5), 
    (20, 8),
]

t = np.linspace(0, 100, 500)

plt.figure(figsize=(10, 7))
plt.quiver(V, P, DV, DP, M, pivot='middle', cmap=plt.cm.viridis)
for V0, P0 in initial_conditions:
    sol = odeint(lotka_volterra, [V0, P0], t)
    plt.plot(sol[:, 0], sol[:, 1], label=f'$V_0={V0}, P_0={P0}$')
plt.scatter(V_s, P_s, color='red', zorder=5, label=f'$V_s = {V_s}, P_s = {P_s}$')
plt.title('Portret fazowy z przykładowymi krzywymi')
plt.xlabel('Liczba ofiar (P)')
plt.ylabel('Liczba drapieżników (V)')
plt.xlim([0, 25]) 
plt.ylim([0, 12.5])  
plt.grid()
plt.legend()
plt.show()
```

Wyznaczymy teraz konkretną zależność liczby drapieżników od liczby ofiar. Możemy formalnie zapisać:
$$
\frac{\mathrm{d}V}{\mathrm{d}P} = \frac{\frac{\mathrm{d}V}{\mathrm{d}t}}{\frac{\mathrm{d}P}{\mathrm{d}t}} = \frac{V(r - aP)}{P(bV - s)}.
$$
Rozdzielamy zmienne i całkujemy:
$$
\int{\left(b - \frac{s}{V} \right){\mathrm{d}V}}  = \int{\left(\frac{r}{P} - a\right){\mathrm{d}P}}.
$$
Stąd otrzymujemy, że:
$$
bV - s\ln{V} - r\ln{P} + aP = c \hspace{0.2cm} {\mathrm{(constans)}}.
$$

Ponieważ równanie zawiera logarytmy, nie przekształcimy go do postaci jawnej, ale rozwiążemy je przy pomocy metod Eulera i Adamsa-Bashfortha. Dodatkowo przybliżymy trajektorie rozwiązań używając wyżej wyprowadzonego wzoru. Weźmiemy przykładowe warunki początkowe równe $\left(V_0, P_0\right) = (40, 9)$ i przeanalizujemy wykres w zależności od stałych współczynników.

```python
import plotly.graph_objs as go
from dash import Dash, dcc, html
from dash.dependencies import Input, Output

def lotka_volterra(X, r, a, b, s):
    V, P = X
    dVdt = r * V - a * V * P
    dPdt = b * V * P - s * P
    return np.array([dVdt, dPdt])

def adams_bashforth_2(f, X0, t0, t_end, h, params):
    t_values = np.arange(t0, t_end + h, h)
    X_values = np.zeros((len(t_values), len(X0)))
    X_values[0] = X0
    X_values[1] = X_values[0] + h * f(X_values[0], *params)
    for i in range(1, len(t_values) - 1):
        X_values[i+1] = X_values[i] + (3/2) * h * f(X_values[i], *params) - (1/2) * h * f(X_values[i-1], *params)
    return t_values, X_values

def euler_method(f, X0, t0, t_end, h, params):
    t_values = np.arange(t0, t_end + h, h)
    X_values = np.zeros((len(t_values), len(X0)))
    X_values[0] = X0
    for i in range(1, len(t_values)):
        X_values[i] = X_values[i-1] + h * f(X_values[i-1], *params)
    return t_values, X_values

r_init, a_init, b_init, s_init = 0.1, 0.02, 0.01, 0.1
V0, P0 = 40, 9
X0 = [V0, P0]
t0, t_end, h = 0, 100, 0.01

app = Dash(__name__)

app.layout = html.Div([
    dcc.Graph(id='phase-plot'),

    html.Div([
        html.Div([
            html.Label("r (przyrost ofiar)", style={'color': 'white'}),
            html.Div(id='slider-r-value', children=f'{r_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-r', min=0, max=1, step=0.01, value=r_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("a (skuteczność drapieżników)", style={'color': 'white'}),
            html.Div(id='slider-a-value', children=f'{a_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-a', min=0, max=1, step=0.01, value=a_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("b (przyrost drapieżników)", style={'color': 'white'}),
            html.Div(id='slider-b-value', children=f'{b_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-b', min=0, max=1, step=0.01, value=b_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("s (śmiertelność drapieżników)", style={'color': 'white'}),
            html.Div(id='slider-s-value', children=f'{s_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-s', min=0, max=1, step=0.01, value=s_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),
    ], style={'margin': '10px', 'padding': '10px', 'display': 'flex', 'flexWrap': 'wrap', 'justifyContent': 'space-between'}),
])

@app.callback(
    [Output('phase-plot', 'figure'),
     Output('slider-r-value', 'children'),
     Output('slider-a-value', 'children'),
     Output('slider-b-value', 'children'),
     Output('slider-s-value', 'children')],
    [Input('slider-r', 'value'),
     Input('slider-a', 'value'),
     Input('slider-b', 'value'),
     Input('slider-s', 'value')]
)
def update_phase_plot(r, a, b, s):
    t_values_ab2, X_values_ab2 = adams_bashforth_2(lotka_volterra, [V0, P0], t0, t_end, h, (r, a, b, s))
    t_values_euler, X_values_euler = euler_method(lotka_volterra, [V0, P0], t0, t_end, h, (r, a, b, s))

    c = b * V0 - s * np.log(V0) - r * np.log(P0) + a * P0
    V = np.linspace(1e-5, 500, 3000)
    P = np.linspace(1e-5, 500, 3000)
    V, P = np.meshgrid(V, P)
    Z = b * V - s * np.log(V) - r * np.log(P) + a * P - c

    fig, ax = plt.subplots()
    CS = plt.contour(V, P, Z, levels=[0])
    plt.close(fig)

    contour_points = []
    for path in CS.get_paths():  
        contour_points.extend(path.vertices)

    contour_points = np.array(contour_points)
    x_contour = contour_points[:, 0]
    y_contour = contour_points[:, 1]

    phase_plot = go.Figure()

    phase_plot.add_trace(go.Scatter(x=x_contour, y=y_contour, mode='lines', name='Model numeryczny', line=dict(color='blue')))
    phase_plot.add_trace(go.Scatter(x=X_values_ab2[:, 0], y=X_values_ab2[:, 1], mode='lines', name='AB2', line=dict(color='green')))
    phase_plot.add_trace(go.Scatter(x=X_values_euler[:, 0], y=X_values_euler[:, 1], mode='lines', name='Euler', line=dict(color='purple', dash='dash')))

    phase_plot.add_trace(go.Scatter(x=[s/b], y=[r/a], mode='markers', name='Punkt stacjonarny', marker=dict(color='black', size=10)))
    phase_plot.update_layout(title="Wykres fazowy: Ofiary (V) vs Drapieżniki (P)", xaxis_title="Liczba ofiar (V)", yaxis_title="Liczba drapieżników (P)", template="plotly_white")
    return phase_plot, f'{r:.2f}', f'{a:.2f}', f'{b:.2f}', f'{s:.2f}'

if __name__ == '__main__':
    app.run_server(debug=False, port=8051)
```

W zależności od wartości parametrów w modelu drapieżnik-ofiara, możemy zaobserwować różne scenariusze, które kształtują dynamikę interakcji między obiema populacjami:
* **Cykliczne oscylacje** - dynamiczne wzajemne oddziaływanie między populacjami, gdzie liczby drapieżników i ofiar oscylują wokół punktu stacjonarnego.
* **Stan równowagi w punkcie przecięcia izoklin** - obie grupy utrzymują się na stabilnym poziomie, w którym wzajemne oddziaływania między nimi nie prowadzą do dalszych zmian liczebności, procesy rozmnażania i śmiertelności równoważą się.
* **Wyginięcie jednej populacji**
    * wyginięcie ofiar - drapieżniki eliminują ofiary szybciej niż te mogą się rozmnażać, co powoduje, że później również giną z powodu braku pożywienia,
    * wyginięcie drapieżników - drapieżniki nie są dosyć skuteczne i wymierają, przez co ofiary mogą się swobodnie rozmnażać.

### Porównanie metod numerycznych

Już na pierwszy rzut oka widać, że metoda Adamsa-Bashfortha przybliża rozwiązania dokładniej niż metoda Eulera. Jednak przeanalizujemy to teraz bardziej szczegółowo. Aby porównać dokładność metod numerycznych dla układu Lotki-Volterry, analizujemy ich błędy dla różnych wartości kroku czasowego $\Delta t$, aby sprawdzić, jak dokładność metod zmienia się w zależności od tego parametru. Błąd dla każdej z metod ($\epsilon$) obliczamy na podstawie różnicy między polem powierzchni zamkniętej trajektorii uzyskanej metodą numeryczną dla jednego pełnego okresu cyklu a wartością dokładną, wyznaczoną za pomocą rozwiązania analitycznego.



```python
from sklearn.linear_model import LinearRegression

def area_gauss(points):
    n = len(points)
    pole = 0
    for i in range(n):
        x, y = points[i-1]  
        xi, yi = points[i]
        xii, yii = points[(i + 1) % n]
        pole += xi * (yii - y)
    return abs(pole) / 2

def find_period(t_values, X_values, X0, difference):
    distances_from_initial = np.linalg.norm(X_values - X0, axis=1)
    min_distance = min(distances_from_initial[2:])
    for i in range(1, len(distances_from_initial)):
        #Sprawdzamy, kiedy odległość zaczyna maleć (zaczynamy zbliżać się do punktu początkowego)
        if distances_from_initial[i] < distances_from_initial[i-1] :
            distance_diff_1 = np.abs(distances_from_initial[i] - min_distance)
            distance_diff_2 = np.abs(distances_from_initial[i+1] - min_distance)
            if distance_diff_1 < difference and distance_diff_1 < distance_diff_2:
                return t_values[i]
    return "Nie udało się znaleźć jednego pełnego okresu (zmień parametry)"

r, a, b, s = 0.8, 0.07, 0.02, 0.5
V0, P0 = 20, 4
X0 = [V0, P0]
t0, t_end, h = 0, 100, 0.01
difference = 100*h
c = b * V0 - s * np.log(V0) - r * np.log(P0) + a * P0

#Znajdujemy rozwiązanie ze wzoru
V = np.linspace(1e-8, 500, 10000)
P = np.linspace(1e-8, 500, 10000)
V, P = np.meshgrid(V, P)
Z = b * V - s * np.log(V) - r * np.log(P) + a * P - c
fig, ax = plt.subplots()
CS = plt.contour(V, P, Z, levels=[0])
plt.close(fig)

contour_points = []
for path in CS.get_paths():  
    contour_points.extend(path.vertices)

contour_points = np.array(contour_points)
area_formula = area_gauss(contour_points)
euler_epsilon = []
ab_epsilon = []
time = []

for i in range(4, 9):
    h = 2**(-1*i)
    time.append(1/h)

    t_values_euler, X_values_euler = euler_method(lotka_volterra, X0, t0, t_end, h, (r, a, b, s))
    period_euler = find_period(t_values_euler, X_values_euler,X0, difference)
    t_values_euler, X_values_euler = euler_method(lotka_volterra, X0, t0, period_euler, h, (r, a, b, s))

    t_values_ab, X_values_ab = adams_bashforth_2(lotka_volterra, X0, t0, t_end, h, (r, a, b, s))
    period_ab = find_period(t_values_ab, X_values_ab,X0, difference)
    t_values_ab, X_values_ab = adams_bashforth_2(lotka_volterra, X0, t0, period_ab, h, (r, a, b, s))
    
    area_euler = area_gauss(X_values_euler)
    epsilon_euler = np.abs(area_euler - area_formula)
    euler_epsilon.append(epsilon_euler)
    
    area_ab = area_gauss(X_values_ab)
    epsilon_ab = np.abs(area_ab - area_formula)
    ab_epsilon.append(epsilon_ab)
    

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
axes[0].scatter(time, euler_epsilon, color='purple', label="Euler")
axes[0].scatter(time, ab_epsilon, color='green', label="Adams-Bashforth 2")
axes[0].legend()
axes[0].set_xlabel(r"Krok czasowy $\frac{1}{\Delta t}$")
axes[0].set_ylabel(r"Błąd $\epsilon$")
axes[0].set_title("Błędy metod numerycznych w zależności od kroku czasowego")

time = np.array(time).reshape(-1, 1)
euler_epsilon = np.array(euler_epsilon)
ab_epsilon = np.array(ab_epsilon)
log_time = np.log(time)
log_euler_epsilon = np.log(euler_epsilon).reshape(-1, 1)
log_ab_epsilon = np.log(ab_epsilon).reshape(-1, 1)
model_euler = LinearRegression().fit(log_time, log_euler_epsilon)
model_a = LinearRegression().fit(log_time, log_ab_epsilon)

axes[1].plot(time, np.exp(model_euler.predict(log_time)), color="purple", linestyle="--")
axes[1].plot(time, np.exp(model_a.predict(log_time)), color="green", linestyle="--")
axes[1].scatter(time, euler_epsilon, color='purple', label="Euler")
axes[1].scatter(time, ab_epsilon, color='green', label="Adams-Bashforth 2")
axes[1].set_xscale("log")
axes[1].set_yscale("log")
axes[1].legend()
axes[1].set_xlabel(r"Krok czasowy  $\frac{1}{\Delta t}$")
axes[1].set_ylabel(r"Błąd $\epsilon$")
axes[1].set_title(f"Błędy metod numerycznych w zależności \n od kroku czasowego (skala log-log)")
plt.show()

print(f'Rząd zbieżności metoda Eulera: {round(np.abs(model_euler.coef_[0][0]))}')
print(f'Rząd zbieżności metoda Adamsa-Bashfortha 2: {round(np.abs(model_a.coef_[0][0]))}')


```

Porównując wyniki obu metod, widać, że metoda Adamsa-Bashfortha jest bardziej dokładna i szybciej zbiega do rozwiązania. Dla mniejszych wartości kroku czasowego błąd metody Eulera jest znacznie większy, podczas gdy metoda Adamsa-Bashfortha oferuje bardziej precyzyjne wyniki nawet przy większych krokach czasowych. Z tego powodu metoda Adamsa-Bashfortha jest bardziej efektywna w przypadku konieczności uzyskania dokładnych wyników w krótszym czasie obliczeniowym.
Dla dokładniejszej analizy błędów, przeprowadzamy regresję log-log na wykresie, gdzie analizujemy zależność logarytmu błędu od logarytmu odwrotności kroku czasowego $\frac{1}{\Delta t}$. \
Ta metoda pozwala wyznaczyć rząd zbieżności obu metod numerycznych:
* Metoda Eulera: rząd zbieżności równy 1, co oznacza, że błąd zmniejsza się w przybliżeniu wprost proporcjonalnie do kroku czasowego.
* Metoda Adamsa-Bashfortha: rząd zbieżności równy 2, co oznacza, że  błąd zmniejsza się dwa razy szybciej niż zmniejsza się krok czasowy. 

Z tego powodu w dalszej analizie modelu drapieżnik-ofiara będziemy skupiać się wyłącznie na metodzie Adama-Bashfortha. 
### Ograniczenia modelu Lotki-Volterry
Chociaż model Lotki-Voltery jest dobrym odzwiercieniem zjawiska oscylacji liczebności drapieżników i ofiar w przyrodzie, to przez swoje uproszczone założenia ma widoczne ograniczenia:
* **Brak uwzględnienia czynników środowiskowych** - model zakłada nieograniczoną przestrzeń i zasoby dla populacji, co w praktyce jest rzadko spotykane; w rzeczywistości zasoby są ograniczone, a przestrzeń może ulegać zmianom,

* **Stałe współczynniki** - parametry takie jak współczynnik przyrostu ofiar (r), współczynnik przyrostu drapieżników (b), współczynnik śmierci drapieżników (s) i współczynnik umierania ofiar (a) są traktowane jako stałe; w rzeczywistości te wartości mogą się zmieniać w czasie w zależności od zmiennych czynników środowiskowych, takich jak dostępność pokarmu, zmiany klimatyczne, sezonowość,

* **Brak konkurencji wewnątrzgatunkowej** - model nie uwzględnia konkurencji między osobnikami tego samego gatunku o zasoby, co może znacząco wpłynąć na dynamikę populacji.


### Rozszerzenia modelu
W odpowiedzi na te ograniczenia zostały opracowane różne rozszerzenia modelu [5]:

* **Pojemność środowiska** - wprowadzenie równania logistycznego, które uwzględnia ograniczone zasoby:
$$
\begin{cases}
    \frac{\mathrm{d}V}{\mathrm{d}t} = rV(1-\frac{V}{K}) - aVP,\\[5pt]
    \frac{\mathrm{d}P}{\mathrm{d}t}= -sP + bVP,
\end{cases}
$$

gdzie *K* - pojemność środowiska. \
Ten ulepszony model przedstawiony jest na poniższym wykresie:

```python
def lotka_volterra_logistic_s(X, r, K, a, b, s):
    V, P = X
    dVdt = r * V * (1 - V / K) - a * V * P 
    dPdt = -s * P + b * V * P 
    return np.array([dVdt, dPdt])

def euler_s(f, X0, t0, t_end, h, r, K, a, b, s):
    t_values = np.arange(t0, t_end + h, h)
    X_values = np.zeros((len(t_values), len(X0)))
    X_values[0] = X0

    for i in range(1, len(t_values)):
        X_values[i] = X_values[i-1] + h * f(X_values[i-1], r, K, a, b, s)
    
    return t_values, X_values

def adams_bashforth_2_s(f, X0, t0, t_end, h, r, K, a, b, s):
    t_values = np.arange(t0, t_end + h, h)
    X_values = np.zeros((len(t_values), len(X0)))
    X_values[0] = X0

    t_values_euler, X_values_euler = euler_s(f, X0, t0, t_end, h, r, K, a, b, s)
    X_values[1] = X_values_euler[1] 

    for i in range(2, len(t_values)):
        f1 = f(X_values[i-1], r, K, a, b, s)
        f2 = f(X_values[i-2], r, K, a, b, s)
        X_values[i] = X_values[i-1] + h / 2 * (3 * f1 - f2)
    
    return t_values, X_values

r_init, K_init, a_init, b_init, s_init = 0.3, 100, 0.1, 0.02, 0.1
V0, P0 = 40, 9
X0 = [V0, P0]
t0, t_end, h = 0, 1000, 0.01

app = Dash(__name__)

app.layout = html.Div([
    html.Div([
        dcc.Graph(id='phase-plot', style={'width': '48%', 'display': 'inline-block'}),
        dcc.Graph(id='time-plot', style={'width': '48%', 'display': 'inline-block'}),
    ], style={'display': 'flex', 'justify-content': 'space-between', 'margin-bottom': '20px'}),

    html.Div([
        html.Div([
            html.Label("r (przyrost ofiar)", style={'color': 'white'}),
            html.Div(id='slider-r-value', children=f'{r_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-r', min=0, max=1, step=0.01, value=r_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("a (skuteczność drapieżników)", style={'color': 'white'}),
            html.Div(id='slider-a-value', children=f'{a_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-a', min=0, max=1, step=0.01, value=a_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("b (przyrost drapieżników)", style={'color': 'white'}),
            html.Div(id='slider-b-value', children=f'{b_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-b', min=0, max=1, step=0.01, value=b_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("s (śmiertelność drapieżników)", style={'color': 'white'}),
            html.Div(id='slider-s-value', children=f'{s_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-s', min=0, max=1, step=0.01, value=s_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
        html.Label("K (pojemność środowiska)", style={'color': 'white'}),
        html.Div(id='slider-K-value', children=f'{K_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
        dcc.Slider(id='slider-K', min=0, max=200, step=20, value=K_init,
                   marks={i * 20: f'{i * 20}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),
    ], style={'margin': '10px', 'padding': '10px', 'display': 'flex', 'flexWrap': 'wrap', 'justifyContent': 'space-between'}),
])

@app.callback(
    [Output('phase-plot', 'figure'),
     Output('time-plot', 'figure'),
     Output('slider-r-value', 'children'),
     Output('slider-K-value', 'children'),
     Output('slider-a-value', 'children'),
     Output('slider-b-value', 'children'),
     Output('slider-s-value', 'children')],
    [Input('slider-r', 'value'),
     Input('slider-K', 'value'),
     Input('slider-a', 'value'),
     Input('slider-b', 'value'),
     Input('slider-s', 'value')]
)
def update_plots(r, K, a, b, s):
    t_values_ab2, X_values_ab2 = adams_bashforth_2_s(lotka_volterra_logistic_s, X0, t0, t_end, h, r, K, a, b, s)

    phase_plot = go.Figure()
    phase_plot.add_trace(go.Scatter(x=X_values_ab2[:, 0], y=X_values_ab2[:, 1], mode='lines', name='AB2', line=dict(color='green')))
    phase_plot.add_trace(go.Scatter(x=[s/b], y=[r/a], mode='markers', name='Punkt równowagi', marker=dict(color='black', size=10)))
    phase_plot.update_layout(title="Wykres fazowy: Ofiary vs Drapieżniki",
                             xaxis_title="Liczba ofiar (V)", yaxis_title="Liczba drapieżników (P)", template="plotly_white")

    time_plot = go.Figure()
    time_plot.add_trace(go.Scatter(x=t_values_ab2, y=X_values_ab2[:, 0], mode='lines', name='Ofiary (AB2)', line=dict(color='blue')))
    time_plot.add_trace(go.Scatter(x=t_values_ab2, y=X_values_ab2[:, 1], mode='lines', name='Drapieżniki (AB2)', line=dict(color='red')))
    time_plot.update_layout(title="Wykres czasowy: Populacja w czasie",
                            xaxis_title="Czas", yaxis_title="Populacja", template="plotly_white")

    return phase_plot, time_plot, f'{r:.2f}', f'{K:.2f}', f'{a:.2f}', f'{b:.2f}', f'{s:.2f}'

if __name__ == '__main__':
    app.run_server(debug=False, port=8053)
```

* **Efekt saturacji drapieżników** - ograniczenie efektywności drapieżnictwa przy niskiej liczbie ofiar lub osiągnięcie maksymalnego tempa polowań przy dużej liczbie ofiar; wynika to z fizycznych i behawioralnych ograniczeń drapieżników:
$$
\frac{\mathrm{d}V}{\mathrm{d}t} = rV- \frac{aVP}{V+D},
$$

gdzie D - stała uwzględniająca efekt saturacji drapieżników, wpływa na efektywność polowań przy małej liczbie ofiar

* **Zależność wzrostu drapieżników od liczebności ofiar** - wzrost populacji drapieżników zależy od dostępności ofiar — ich liczebność ogranicza tempo reprodukcji i przetrwanie drapieżników; przy niskiej liczbie ofiar populacja drapieżników maleje.
$$
\frac{\mathrm{d}P}{\mathrm{d}t}= -sP + bP(1-\frac{HP}{V}),
$$

gdzie H - współczynnik określający zależność wzrostu populacji drapieżników od liczebności ofiar (im wyższe, tym silniejszy wpływ niedoboru 
ofiar na drapieżniki.)

Poniżej przedstawione zostały wykresy z dodanym efektem saturacji i zależnością wzrostu drapieżników od liczebności ofiar:

```python
def lotka_volterra_s(X, r, a, b, H, D, s):
    V, P = X
    V = max(V, 0)
    P = max(P, 0)
    
    dVdt = r * V * (1 - V / 1000) - (a * V * P) / (V + D)
    dPdt = b * P * (1 - (H * P) / max(V, 1e-6)) - s * P
    return np.array([dVdt, dPdt])

def euler_s(f, X0, t0, t_end, h, r, a, b, s, H, D):
    t_values = np.arange(t0, t_end + h, h)
    X_values = np.zeros((len(t_values), len(X0)))
    X_values[0] = X0

    for i in range(1, len(t_values)):
        X_values[i] = X_values[i-1] + h * f(X_values[i-1], r, a, b, s, H, D)
    
    return t_values, X_values

def adams_bashforth_2_s(f, X0, t0, t_end, h, r, a, b, s, H, D):
    t_values = np.arange(t0, t_end + h, h)
    X_values = np.zeros((len(t_values), len(X0)))
    X_values[0] = X0

    t_values_euler, X_values_euler = euler_s(f, X0, t0, t_end, h, r, a, b, s, H, D)
    X_values[1] = X_values_euler[1]  

    for i in range(2, len(t_values)):
        f1 = f(X_values[i-1], r, a, b, s, H, D)
        f2 = f(X_values[i-2], r, a, b, s, H, D)
        X_values[i] = X_values[i-1] + h / 2 * (3 * f1 - f2)
    
    return t_values, X_values

r_init, a_init, H_init, D_init, b_init, s_init = 0.3, 0.1, 1.3, 0.2, 0.4, 0.2
V0, P0 = 40, 9
X0 = [V0, P0]
t0, t_end, h = 0, 200, 0.01

app = Dash(__name__)

app.layout = html.Div([
    html.Div([
        dcc.Graph(id='phase-plot', style={'width': '48%', 'display': 'inline-block'}),
        dcc.Graph(id='time-plot', style={'width': '48%', 'display': 'inline-block'}),
    ], style={'display': 'flex', 'justify-content': 'space-between', 'margin-bottom': '20px'}),

    html.Div([
        html.Div([
            html.Label("r (przyrost ofiar)", style={'color': 'white'}),
            html.Div(id='slider-r-value', children=f'{r_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-r', min=0, max=1, step=0.01, value=r_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("a (skuteczność drapieżników)", style={'color': 'white'}),
            html.Div(id='slider-a-value', children=f'{a_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-a', min=0.01, max=1, step=0.01, value=a_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(1, 11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("b (przyrost drapieżników)", style={'color': 'white'}),
            html.Div(id='slider-b-value', children=f'{b_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-b', min=0, max=1, step=0.01, value=b_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("s (śmiertelność drapieżników)", style={'color': 'white'}),
            html.Div(id='slider-s-value', children=f'{s_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-s', min=0, max=1, step=0.01, value=s_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("H (czynnik drapieżnictwa)", style={'color': 'white'}),
            html.Div(id='slider-H-value', children=f'{H_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-H', min=0.25, max=5, step=0.01, value=H_init,
                       marks={i: f'{i:.1f}' for i in np.arange(0, 5.5, 0.5)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

        html.Div([
            html.Label("D (wpływ drapieżników na ofiary)", style={'color': 'white'}),
            html.Div(id='slider-D-value', children=f'{D_init:.2f}', style={'textAlign': 'center', 'color': 'white'}),
            dcc.Slider(id='slider-D', min=0.1, max=1, step=0.01, value=D_init,
                       marks={i / 10: f'{i / 10:.1f}' for i in range(11)}),
        ], style={'width': '48%', 'display': 'inline-block', 'margin': '10px'}),

    ], style={'margin': '10px', 'padding': '10px', 'display': 'flex', 'flexWrap': 'wrap', 'justifyContent': 'space-between'}),

])

@app.callback(
    [Output('phase-plot', 'figure'),
     Output('time-plot', 'figure'),
     Output('slider-r-value', 'children'),
     Output('slider-a-value', 'children'),
     Output('slider-b-value', 'children'),
     Output('slider-s-value', 'children'),
     Output('slider-H-value', 'children'),
     Output('slider-D-value', 'children')],
    [Input('slider-r', 'value'),
     Input('slider-a', 'value'),
     Input('slider-b', 'value'),
     Input('slider-s', 'value'),
     Input('slider-H', 'value'),
     Input('slider-D', 'value')]
)
def update_plots(r, a, b, s, H, D):
    t_values_ab2, X_values_ab2 = adams_bashforth_2_s(lotka_volterra_s, X0, t0, t_end, h, r, a, b, s, H, D)

    phase_plot = go.Figure()
    phase_plot.add_trace(go.Scatter(x=X_values_ab2[:, 0], y=X_values_ab2[:, 1], mode='lines', name='AB2', line=dict(color='green')))
    phase_plot.update_layout(title="Wykres fazowy: Ofiary vs Drapieżniki",
                             xaxis_title="Liczba ofiar (V)", yaxis_title="Liczba drapieżników (P)", template="plotly_white")

    time_plot = go.Figure()
    time_plot.add_trace(go.Scatter(x=t_values_ab2, y=X_values_ab2[:, 0], mode='lines', name='Ofiary (AB2)', line=dict(color='blue')))
    time_plot.add_trace(go.Scatter(x=t_values_ab2, y=X_values_ab2[:, 1], mode='lines', name='Drapieżniki (AB2)', line=dict(color='red')))
    time_plot.update_layout(title="Wykres czasowy: Populacja w czasie",
                            xaxis_title="Czas", yaxis_title="Populacja", template="plotly_white")

    return phase_plot, time_plot, f'{r:.2f}', f'{a:.2f}', f'{b:.2f}', f'{s:.2f}', f'{H:.2f}', f'{D:.2f}'

if __name__ == '__main__':
    app.run_server(debug=False, port=8054)
```

### Przykłady zastosowań

Model i jego rozszerzenia znajduje szerokie zastosowanie w różnych dziedzinach:

* **Ekologia:** Przykład relacji między populacjami wilków i zajęcy w ekosystemach borealnych jest klasycznym przypadkiem zastosowania modelu, gdzie drapieżnik i ofiara wpływają na siebie nawzajem w cykliczny sposób.

* **Ekonomia:** W ekonomii model ten bywa wykorzystywany do opisu konkurencji między przedsiębiorstwami. Przykładowo, firmy mogą konkurować o ograniczone zasoby rynku, co prowadzi do wzrostu jednej firmy kosztem drugiej.

* **Epidemiologia:** Model może być używany do badania interakcji między wirusem a układem odpornościowym gospodarza, w którym wirus działa jak drapieżnik, a organizm jest ofiarą. Może również opisywać inne zjawiska epidemiologiczne, takie jak zmieniająca się liczba przypadków chorób zakaźnych.

### Dane historyczne
Przedstawiony poniżej wykres przedstawia dane historyczne pochodzące od "Kampanii Zatoki Hudsona", która od 1670 roku prowadzi handel futrami w Kanadzie [4]. Na przestrzeni lat zbierała ona informacje na temat liczebności rysia kanadyjskiego i zająca amerykańskiego, który jest głównym pożywieniem rysi w północnej tajdze Ameryki Północnej. Populacje obu gatunków są ściśle ze sobą powiązane, co pozwala zaobserwować charakterystyczne cykle populacyjne. Te dane są doskonałym przykładem zależności opisanych w modelu drapieżnik-ofiara. [3]


```python
import requests

url = 'http://people.whitman.edu/~hundledr/courses/M250F03/LynxHare.txt'
response = requests.get(url)
data = response.text.splitlines()

years = []
hare = []
lynx = []

for line in data:
    values = line.split()
    years.append(float(values[0]))  
    hare.append(float(values[1]))   
    lynx.append(float(values[2]))  

plt.figure(figsize=(10, 6))
plt.plot(years, hare, label='Zające', color='blue')
plt.plot(years, lynx, label='Lisy', color='red')
plt.title('Populacja rysi i zająców w latach 1845-1935')
plt.xlabel('Rok')
plt.ylabel('Populacja (w tysiącach)')
plt.grid(True)
plt.xlim([1845, 1935]) 
plt.ylim([0, 160])
plt.legend()
plt.show()
```

Na podstawie wykresu można zauważyć, że populacje zajęcy i rysi w ekosystemach tajgi wykazują cykle, które powtarzają się co 8–11 lat. Maksimum liczby drapieżników następuje z opóźnieniem 1–2 lat po maksimum liczby ofiar. Dodatkowo wyniki te są kształtowane przez zmieniające się warunki środowiskowe, takie jak dostępność pożywienia dla zajęcy czy zmienne warunki klimatyczne, które mogą nasilać spadki populacji w okresach kryzysu.

Model Lotki-Volterry, mimo swoich uproszczeń, jest użytecznym narzędziem do badania interakcji międzygatunkowych.
